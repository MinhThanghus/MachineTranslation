```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

## **English to Vietnamese**

```python
bi          = [26.52, 29.59, 30.84, 32.06, 32.67, 32.56, 32.97]
finetune_bi = [None , 29.59, 31.24, 33.07, 34.14, 34.90, 34.81, 34.94]
uni         = [32.70, 34.60, 35.26, 35.59, 36.61, 36.80, 36.81, 36.96, 37.48]
google      = [33.01]*8


plt.plot([str(i) for i in range(1, len(bi)+1)], bi)
plt.plot([str(i) for i in range(1, len(finetune_bi)+1)], finetune_bi)
plt.plot([str(i) for i in range(1, len(uni)+1)], uni)
plt.plot([str(i) for i in range(1, len(google)+1)], google)
plt.legend(['Bi-directional', 'Fine Tune for En2Vi', 'Uni-Directional', 'Google Translate'])
plt.xlabel('Train Steps (100k)')
plt.ylabel('BLEU Score')
plt.title('BLEU Score in tst2013 (beamsize=4) - English to Vietnamese')
```

## **Vietnamese to English**

```python
bi          = [21.19, 27.05, 29.01, 30.54, 31.15, 31.91, 31.86]
finetune_bi = [21.19, 29.11, 31.65, 32.33, 32.94]
uni         = [18.14, 26.89, 29.92, 31.38, 32.07]
google      = [27.78]*len(bi)


plt.plot([str(i) for i in range(1, len(bi)+1)], bi)
plt.plot([str(i) for i in range(1, len(finetune_bi)+1)], finetune_bi)
plt.plot([str(i) for i in range(1, len(uni)+1)], uni)
plt.plot([str(i) for i in range(1, len(google)+1)], google)
plt.legend(['Bi-directional', 'Fine-tune for Vi2En',  'Uni-Directional', 'Google Translate'])
plt.xlabel('Train Steps (100k)')
plt.ylabel('BLEU Score')
plt.title('BLEU Score in tst2013 (beamsize=4) - Vietnamese to English')
```
